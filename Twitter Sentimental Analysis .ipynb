```python
import pandas as pd
from textblob import TextBlob as tb 
import matplotlib.pyplot as plt 
import nltk
import re
```

```python
#give the csv path
df = pd.read_csv(r'C:\ClassesMSC\DataSets\tweets.csv')
df.shape
```

```python
df = df.dropna(axis = 0, how ='any')
df.shape
```

```python
df.head(5)
```

```python
df.describe()
```

```python
df['text'] = df['text'].str.lower()
```

```python
def tokenize_text(text):
    tokens = nltk.word_tokenize(text)
    return tokens
```

```python
df['tokenized_text'] = df['text'].apply(tokenize_text)
```

```python
df.head()
```

```python
stop_words = set(nltk.corpus.stopwords.words('english'))
```

```python
def remove_stopwords(text):
    tokens = nltk.word_tokenize(text)
    filtered_tokens = [token for token in tokens if token.lower() not in stop_words]
    filtered_text = ' '.join(filtered_tokens)
    return filtered_text
```

```python
df['text'] = df['text'].apply(remove_stopwords)
```

```python
df.head()
```

```python
# Stemming 
```

```python
from nltk.stem import PorterStemmer
```

```python
stemmer = PorterStemmer()
```

```python
def stem_text(text):
    tokens = nltk.word_tokenize(text)
    stemmed_tokens = [stemmer.stem(token) for token in tokens]
    stemmed_text = ' '.join(stemmed_tokens)
    return stemmed_text
```

```python
df['text'] = df['text'].apply(stem_text)
```

```python
# Lemmatizing 
```

```python
from nltk.stem import WordNetLemmatizer
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
def lemmatize_text(text):
    tokens = nltk.word_tokenize(text)
    lemmatized_tokens = [lemmatizer.lemmatize(token) for token in tokens]
    lemmatized_text = ' '.join(lemmatized_tokens)
    return lemmatized_text
```

```python
df['text'] = df['text'].apply(lemmatize_text)
```

```python
# Removing special characters and numerical values
```

```python
pattern = r'[^a-zA-Z\s]|\d+'
```

```python
def remove_special_chars(text):
    cleaned_text = re.sub(pattern, '', text)
    return cleaned_text
```

```python
df['text'] = df['text'].apply(remove_special_chars)
```

```python
df.head()
```

```python
# Part-of-speech tagging
```

```python
def pos_tagging(text):
    tokens = nltk.word_tokenize(text)
    tagged_tokens = nltk.pos_tag(tokens)
    tagged_text = ' '.join([token[1] for token in tagged_tokens])
    return tagged_text
```

```python
df['tokenized_text'] = df['text'].apply(pos_tagging)
```

```python
df.head(5)
```

```python
df['tokenized_text']
```

```python
polarity = []
subjectivity = []
text = []
NPN = []

for i in df['text']:
    blob1 = tb(i)
    out = blob1.sentiment
    polarity.append(out[0])
    subjectivity.append(out[1])
    text.append(i)
    if out[0] == 0:
        NPN.append("Neu")
    elif out[0] >0:
        NPN.append("Pos")
    else:
        NPN.append("Neg")
```

```python
final =pd.merge(df,pd.DataFrame({"polarity" :polarity,
                                 "subjectivity": subjectivity,
                                 "text": text,
                                 "NPN_p":NPN}), on ="text",how = "inner")

final
```

```python
final['NPN_p']
```

```python
output = final.groupby(final['NPN_p']).count()
```
